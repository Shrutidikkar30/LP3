```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay, precision_score, recall_score, accuracy_score
```

```python
data_frame=pd.read_csv('emails.csv')
```

```python
data_frame
```

```python
data_frame=data_frame.drop(['Email No.'], axis=1)
```

```python
data_frame
```

```python
X = data_frame.drop('Prediction', axis = 1)
Y = data_frame['Prediction']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(X, Y, train_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn_model=KNeighborsClassifier(n_neighbors=2)
kn_model.fit(x_train, y_train)
```

```python
y_pred=kn_model.predict(x_test)
cm=confusion_matrix(y_test, y_pred)
```

```python
mat=ConfusionMatrixDisplay(confusion_matrix=cm)
mat.plot()
#mat.show()
```
